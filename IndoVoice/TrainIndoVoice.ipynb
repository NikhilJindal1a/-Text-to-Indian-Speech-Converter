```python
# Install necessary packages (if not already installed)
!pip install pydub librosa

# Import necessary libraries
from pydub import AudioSegment
from pydub.playback import play
from IPython.display import display, clear_output
import ipywidgets as widgets
import base64
from gtts import gTTS
from io import BytesIO
import librosa
import numpy as np

# Function to handle button click event for text-to-speech conversion
def on_button_clicked(b):
    text = text_input.value.strip()
    if text:
        with output:
            output.clear_output()
            print("Converting text to speech, please wait...")

        # Synthesize text to speech using sample voice
        synthesized_audio = text_to_speech(text)

        # Save the synthesized speech to an MP3 file
        synthesized_audio.export("output.mp3", format="mp3")

        # Read the MP3 file and convert it to base64
        with open("output.mp3", "rb") as audio_file:
            encoded_audio = base64.b64encode(audio_file.read()).decode('utf-8')

        # Display the audio player
        audio_html = f'<audio autoplay controls><source src="data:audio/mpeg;base64,{encoded_audio}" type="audio/mpeg"></audio>'
        clear_output(wait=True)
        display(title, text_input, button, clear_button, output)
        display(widgets.HTML(value=audio_html))

        # Provide feedback to the user
        with output:
            print("Conversion complete. Enter new text for another conversion.")
    else:
        with output:
            output.clear_output()
            print("Please enter some text.")

# Function to synthesize text to speech using the sample voice
def text_to_speech(text):
    global sample_voice

    # Synthesize text to speech using gTTS with the provided sample voice data
    tts = gTTS(text=text, lang='hi', slow=False)
    fp = BytesIO()
    tts.write_to_fp(fp)
    fp.seek(0)
    synthesized_audio = AudioSegment.from_file(fp)

    return synthesized_audio

# Function to handle button click event for clearing text input
def on_clear_clicked(b):
    text_input.value = ""  # Clear text input area

# Function to convert voice characteristics
def convert_voice(audio_data, target_audio):
    # Calculate the length ratio between audio data and target audio
    ratio = len(audio_data) / len(target_audio)

    # Perform time stretching on the audio data
    audio_stretched = librosa.effects.time_stretch(np.array(audio_data.get_array_of_samples()), ratio)

    return AudioSegment(audio_stretched.tobytes(), frame_rate=audio_data.frame_rate, sample_width=audio_data.sample_width, channels=audio_data.channels)

# Path to your target voice data file (replace with your file path)
target_voice_file = "/content/drive/MyDrive/vrarmr/IndoVoice/WhatsApp_Audio.wav"

# Load sample voice data
sample_voice = AudioSegment.from_file(target_voice_file)

# Create text input widget
text_input = widgets.Textarea(
    placeholder='Enter text here',
    description='Text:',
    disabled=False,
    layout=widgets.Layout(width='70%', height='100px')
)

# Create button widget for converting text to speech
button = widgets.Button(
    description="Convert to Speech",
    button_style='success',
    tooltip="Click to convert text to speech",
    icon='microphone'
)

# Create button widget for clearing text input
clear_button = widgets.Button(
    description="Clear Text",
    button_style='warning',
    tooltip="Click to clear the text input",
    icon='trash'
)

# Create an output widget for feedback
output = widgets.Output()

# Assign click event
button.on_click(on_button_clicked)
clear_button.on_click(on_clear_clicked)

# Create a title widget
title = widgets.HTML(value="<h2>Text to Speech Converter</h2>")

# Display widgets
display(title, text_input, button, clear_button, output)
```
